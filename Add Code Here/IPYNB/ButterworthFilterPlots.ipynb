```python
%matplotlib inline
from scipy import signal as sig
from matplotlib.pyplot import *
from math import *
```

# Butterworth filter

```python
# rad/s
w0 = 1000 

# First order filter
num = [1]
den = [1/w0, 1]

tf = sig.TransferFunction(num, den)
w, mag, phase = sig.bode(tf)

semilogx(w, mag, '-')

# Second order filter
den = [1/w0**2, sqrt(2)/w0, 1]

tf = sig.TransferFunction(num, den)
w, mag, phase = sig.bode(tf)

semilogx(w, mag)

grid(which="both")
semilogx(1000, -3, '*b')
```

```python
# High pass

# First order
num = [1]
den = [1/w0, 1]

tf = sig.TransferFunction(num, den)
w, mag, phase = sig.bode(tf)

semilogx(w, mag, '-')

# Second order
den = [1/w0**2, sqrt(2)/w0, 1]

tf = sig.TransferFunction(num, den)
w, mag, phase = sig.bode(tf)

semilogx(w, mag)

# Calculate poles of 3th order filter
num, den = sig.butter(3, w0, btype='low', analog=True, output='ba')
tf = sig.TransferFunction(num, den)
w, mag, phase = sig.bode(tf)

semilogx(w, mag)

num, den = sig.butter(4, w0, btype='low', analog=True, output='ba')
tf = sig.TransferFunction(num, den)
w, mag, phase = sig.bode(tf)

semilogx(w, mag)

grid(which="both")
semilogx(1000, -3, '*b')
```
