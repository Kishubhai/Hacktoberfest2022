# Decision Tree

This task is to build a decision tree diagram. The parameter will be 

\[Age < 21, Income = Low, Gender = Female, Martial Status = Married]

## Import the library

```python
import pandas as pd
import numpy as np
```

## Import the dataset

The dataset can be changed like you want.

If you not find the dataset here

Just rewrite it. Trust me it's worth it.

```python
df = pd.read_csv('data.csv')
df
```

## Search the Buys Parameter

### Using iloc -1

```python
x = df.iloc[:,:-1]
x
```

### Using iloc 5

```python
y = df.iloc[:,5]
y
```

## Import label encoder text for labelling

```python
from sklearn.preprocessing import LabelEncoder
```

```python
x = x.apply(LabelEncoder().fit_transform)
x
```

## Import Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
```

## Make the model

```python
dtree = DecisionTreeClassifier()

dtree.fit(x.iloc[:,1:5], y)
```

```python
x_question = np.array([1, 1, 0, 0])
```

```python
y_pred = dtree.predict([x_question])
y_pred
```

## Visualizing using DT

```python
from sklearn.tree import export_graphviz
from IPython.display import Image
from six import StringIO
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(dtree, 
                out_file = dot_data, 
                filled = True, 
                rounded = True, 
                special_characters = True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('dtree.png')
```
